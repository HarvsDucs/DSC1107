### Summarize

```python
import pandas as pd
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Calculate 'price_per_carat' first (as in the previous step)
diamonds['price_per_carat'] = diamonds['price'] / diamonds['carat']


# Calculate the number of "fancy" diamonds
num_fancy_diamonds = (diamonds['price_per_carat'] > 10000).sum()

# Create a DataFrame for output
result_df = pd.DataFrame({'num_fancy_diamonds': [num_fancy_diamonds]})


# Print the result
print(result_df)
```

```python
diamonds['price_per_carat'] = diamonds['price'] / diamonds['carat']

# Calculate the number of "fancy" diamonds and the mean price
num_fancy_diamonds = (diamonds['price_per_carat'] > 10000).sum()
mean_diamond_price = diamonds['price'].mean()

# Create a DataFrame for the result
result_df = pd.DataFrame({
    'num_fancy_diamonds': [num_fancy_diamonds],
    'mean_diamond_price': [mean_diamond_price]
})

# Print the result
print(result_df)
```

## Multi Step Transformations

### The pipe

```python
import pandas as pd
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Perform the sequence of operations using chaining
processed_diamonds = (
    diamonds
    .query('cut == "Premium"')
    .assign(price_per_carat=lambda x: x['price'] / x['carat'])
    .sort_values(by='price_per_carat', ascending=False)
)

# Print the processed DataFrame (optional)
print(processed_diamonds)
```

### Group By

```python
import pandas as pd
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Group by 'cut' and find the maximum price for each group
max_prices_by_cut = (
    diamonds
    .groupby('cut')
    ['price']
    .max()
    .reset_index(name='max_price')
)

# Print the result
print(max_prices_by_cut)
```

```python
import pandas as pd
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Group by 'cut' and 'clarity' and find the maximum price for each group
max_prices_by_cut_clarity = (
    diamonds
    .groupby(['cut', 'clarity'])
    ['price']
    .max()
    .reset_index(name='max_price')
)

# Print the result
print(max_prices_by_cut_clarity)
```

```python
import pandas as pd
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Group, summarize, and then "ungroup" (reset index)
max_prices_by_cut_clarity = (
    diamonds
    .groupby(['cut', 'clarity'])
    ['price']
    .max()
    .reset_index(name='max_price')
)

# Print the result
print(max_prices_by_cut_clarity)
```
